```python
!pip install kagglehub

import kagglehub
import pandas as pd
import os

# Download the Superstore dataset
path = kagglehub.dataset_download("vivek468/superstore-dataset-final")
print("Downloaded to:", path)

# Find the CSV file inside the downloaded folder
for f in os.listdir(path):
    print("File found:", f)
```

```python
# Load the CSV (replace filename if different)
df = pd.read_csv(path + "/Sample - Superstore.csv", encoding='latin-1')

print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# 1. Fix column names (clean formatting)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("Columns:", df.columns.tolist())

# 2. Check & fix data types
print("\nData Types:\n", df.dtypes)
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])

# 3. Check for outliers in Sales & Profit
print("\nSales stats:\n", df['sales'].describe())
print("\nProfit stats:\n", df['profit'].describe())

# 4. Extract date parts (useful for dashboard)
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_quarter'] = df['order_date'].dt.quarter

# 5. Standardize text columns
df['region'] = df['region'].str.strip().str.title()
df['category'] = df['category'].str.strip().str.title()
df['segment'] = df['segment'].str.strip().str.title()

# 6. Save final clean file
df.to_csv('superstore_final_clean.csv', index=False)
print("\nâ Final clean file saved! Shape:", df.shape)
```

```python
# Run this in Colab to get a summary of your clean data
print("=== DASHBOARD PLANNING SUMMARY ===")
print("\nDate Range:", df['order_date'].min(), "to", df['order_date'].max())
print("\nTotal Sales: $", round(df['sales'].sum(), 2))
print("\nTotal Profit: $", round(df['profit'].sum(), 2))
print("\nCategories:\n", df['category'].value_counts())
print("\nRegions:\n", df['region'].value_counts())
print("\nSegments:\n", df['segment'].value_counts())
print("\nTop 5 Sub-Categories by Sales:\n", df.groupby('sub-category')['sales'].sum().sort_values(ascending=False).head())
```
